 January 2020 data presents the ‘baseline’ of what the NYC Taxi business used to be.  
• Compare the data of Jan-2020 vs Mar-2020. 

```python
 df_jan= pd.read_parquet('yellow_tripdata_2020-01.parquet', engine='fastparquet') 
 
 df_mar= pd.read_parquet('yellow_tripdata_2020-03.parquet', engine='fastparquet') 
```

```python
# Function to clean data
def clean_data(df):
    # Select relevant columns
    df = df[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'passenger_count', 'trip_distance', 'fare_amount', 'total_amount']]
    
    # Convert datetime columns
    df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
    df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
    
    # Remove trips with zero passengers or zero fare
    df = df[(df['passenger_count'] > 0) & (df['fare_amount'] > 0)]
    
    return df

df_jan = clean_data(df_jan)
df_mar= clean_data(df_mar)
```

Shows key ride statistics before and after the pandemic.

```python
# Summary statistics for January
summary_jan = df_jan[['passenger_count', 'trip_distance', 'fare_amount', 'total_amount']].describe()

# Summary statistics for March
summary_mar = df_mar[['passenger_count', 'trip_distance', 'fare_amount', 'total_amount']].describe()

# Print comparison
print("Summary - January 2020 (Pre-COVID):\n", summary_jan)
print("Summary - March 2020 (Post-COVID):\n", summary_mar)
```

-Shows the drastic drop in taxi rides after mid-March 2020.

```python
import matplotlib.pyplot as plt

df_jan['date'] = df_jan['tpep_pickup_datetime'].dt.date
df_mar['date'] = df_mar['tpep_pickup_datetime'].dt.date

# Group by date
trips_per_day_jan = df_jan.groupby('date').size()
trips_per_day_mar = df_mar.groupby('date').size()

# Plot comparison
plt.figure(figsize=(12, 6))
plt.plot(trips_per_day_jan, label="January 2020", marker="o")
plt.plot(trips_per_day_mar, label="March 2020", marker="o", linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("Number of Trips")
plt.title("NYC Taxi Trips: Jan 2020 vs. Mar 2020")
plt.legend()
plt.show()
```

Trip Distance & Fare Amount Comparison
Indicates whether people took shorter or longer trips post-COVID.


```python
import seaborn as sns

plt.figure(figsize=(12, 5))

# Trip Distance
plt.subplot(1, 2, 1)
sns.histplot(df_jan['trip_distance'], bins=50, color="blue", label="January 2020", kde=True)
sns.histplot(df_mar['trip_distance'], bins=50, color="red", label="March 2020", kde=True)
plt.xlabel("Trip Distance (miles)")
plt.title("Trip Distance Distribution")
plt.legend()

# Fare Amount
plt.subplot(1, 2, 2)
sns.histplot(df_jan['fare_amount'], bins=50, color="blue", label="January 2020", kde=True)
sns.histplot(df_mar['fare_amount'], bins=50, color="red", label="March 2020", kde=True)
plt.xlabel("Fare Amount ($)")
plt.title("Fare Amount Distribution")
plt.legend()

plt.tight_layout()
plt.show()
```
